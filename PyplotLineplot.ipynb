```
import pandas as pd
from google.colab import files
uploaded = files.upload()
import io
df2 = pd.read_csv(io.BytesIO(uploaded['istanbul.csv']))
//rint(df2)
```

Line plot işlemleri için matplotlib.pyplot kütüphanesini import ediyoruz.

```
import matplotlib.pyplot as plt
```

```
private_room = df2[df2.room_type == "Private room"]
shared_room = df2[df2.room_type == "Shared room"]
entire_room = df2[df2.room_type == "Entire home/apt"]
```

pyplot kütüphanesinin plot fonksiyonunu çağırıyoruz. 
ilk parametresi = x eksenidir.
ikinci parametresi = y eksenidir.
x ve y yazmaya gerek yoktur. Sadece iki parametre ile fonksiyon çağırılabilir.
Aşağıda temel olarak verilen bir plot fonksiyonunda olması gereken parametreler gösterilmiştir.

```
df2head = private_room.head(15) # data çok olduğu için ilk 15 kayıt üzerinden bakalım
plt.plot(df2head.extra_people,df2head.price,color = "red",label = "private room - extra people")
plt.show()
```

plot fonksiyonun linestyle parametresi kullanılarak çizim style'i değiştirebilir. 

linestyle e aşağıdakilerden biri verilebilir.
*  ':'	dotted line style
*   '.'	point marker
*   ','	pixel marker
*   'o'	circle marker

```
plt.plot(df2head.extra_people,df2head.price,color = "red",label = "private room - extra people",linestyle =":")
plt.show()
```

Plot fonksiyonuna alpha parametresi verilerek çizimin saydamlığı ayarlanabilir.
Eğer 0'a yakın bir değer verilirse oldukça saydam olur. **0'da görülmez.**

```
plt.plot(df2head.extra_people,df2head.price,color = "red",label = "private room - extra people",alpha =0.1)
plt.show()
```

x ve y ekseninin label etiketlerinin ne olduğu belli değildir. Bunlara etiketleri verilebilir.

```
plt.plot(df2head.extra_people,df2head.price,color = "red",label = "private room - extra people")
plt.xlabel("Extra People")
plt.ylabel("Price")
plt.show()
```

Bu çizime başlık ekleyelim. En üstte yazacaktır.

```
plt.plot(df2head.extra_people,df2head.price,color = "red",label = "private room - extra people")
plt.title("Extra People - Price")
plt.xlabel("Extra People")
plt.ylabel("Price")
plt.show()
```

bir tane sorun var. plot fonksiyonuna verdiğimiz label görünmemektedir. Bunun görünmesi için pyplot kütüphanesinin legend() fonksiyonunu çağırıyoruz.  Eğer parametre verilmezse en uygun yere yerleştilecektir.

```
plt.plot(df2head.extra_people,df2head.price,color = "red",label = "private room - extra people")
plt.legend()
plt.show()
```

label etiketinin konumunun ayarlanması için:
loc parametresi kullanılır. loc parametresinin alabileceği değerler : 


*   best
*   upper right
*   upper left
*   lower left
*   lower right
*   right
*   center left
*   center  right
*   lower center
*   upper  center
*   center

```
plt.plot(df2head.extra_people,df2head.price,color = "red",label = "private room - extra people")
plt.legend(loc="center")
plt.show()
```

**Not: **  plot fonksiyonundan sonra çizimi görmek istiyorsan **pyplot kütüphanesi**nin **show()** fonksiyonunu çağırmak gerekiyor.
